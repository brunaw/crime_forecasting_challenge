# Evolution by date

```python
# converting to date format
crimes['occ_date'] = pd.to_datetime(crimes['occ_date'], format='%Y-%m-%d')
```

```python
# dropping coordinates
evolution = crimes[['CATEGORY','occ_date']]
evolution.head()
```

## Sorting by date

```python
# sorting by date
evolution = evolution.sort_values(by='occ_date')
evolution.head()
```

### Adding date to index

```python
evolution.index = evolution['occ_date']
```

### Grouping by day

```python
# grouping and counting crimes on each day
counts_day =evolution.groupby(['occ_date']).agg(['count'])
counts_day.head()
```

### Grouping by week

```python
# grouping by week
counts_week = evolution.groupby(pd.TimeGrouper(freq='W')).agg(['count'])
counts_week = counts_week.drop('occ_date', axis=1)
counts_week.head()
```

```python
counts_week_pivot = pd.pivot_table(counts_week, index=counts_week.index.week, columns=counts_week.index.year, values='CATEGORY', aggfunc='sum')
counts_week_pivot = counts_week_pivot.fillna(counts_week_pivot.mean())
counts_week_pivot.head()
```

```python
counts_week_pivot.plot(figsize=(50,10))
plt.xticks([n for n in range(1,52)])
plt.savefig('../../plots/burglary_week_plot.png')
plt.show()
```

```python
counts_week_april = evolution.loc[evolution.index.month == 4].groupby(pd.TimeGrouper(freq='D')).agg(['count'])
counts_week_april = counts_week_april.drop('occ_date', axis=1)
counts_week_april.head()
```

```python
counts_week_april_pivot = pd.pivot_table(counts_week_april, index=counts_week_april.index.day, columns=counts_week_april.index.year, values='CATEGORY', aggfunc='sum')
#counts_week_april_pivot = counts_week_april_pivot[13:17]
counts_week_april_pivot = counts_week_april_pivot.fillna(counts_week_april_pivot.mean())
counts_week_april_pivot.head()
```

### Plotting mean for every year in April

```python
#counts_week_april_pivot[:30].plot(figsize=(20,10))
#print(counts_week_april_pivot.mean(axis=1))
plt.plot(counts_week_april_pivot.mean(axis=1)[:30])
plt.xticks([n for n in range(1,31)])
plt.savefig('../../plots/burglary_week_april_plot.png')
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.plot(counts_week_pivot.mean(axis=1))
plt.xticks([n for n in range(1,52)])
plt.savefig('../../plots/burglary_week_mean_plot.png')
plt.show()
```

### Grouping by month

```python
counts_month = evolution.groupby(pd.TimeGrouper(freq='M')).agg(['count'])

counts_month = counts_month.drop('occ_date', axis=1)
counts_month.head()
```

## Pivot transformation

```python
counts_month_pivot = pd.pivot_table(counts_month, index=counts_month.index.month, columns=counts_month.index.year, values='CATEGORY', aggfunc='sum')
counts_month_pivot.head()
```

```python
# Month numbers to literal
import calendar
counts_month_pivot = counts_month_pivot.rename(index=lambda x:  calendar.month_abbr[x])
```

```python
# Modifying the legends
ax1 = counts_month_pivot.plot.bar(figsize=(20,10), use_index=True, grid=True, sort_columns=True)
lines, labels = ax1.get_legend_handles_labels()
print(labels)
new_labels = []
for l in labels:
    new_labels.append(l[8:len(l)-1])
print(new_labels)
ax1.legend(lines, new_labels, loc='best')  # improved legends

plt.savefig('../../plots/burglary_month_plot_barh.png')
plt.show()
```

```python
# plot
ax2 = counts_month_pivot.plot(figsize=(20,10), use_index=True, grid=True)
ax2.legend(lines, new_labels, loc='best')  # improved legends
plt.savefig('../../plots/burglary_month_plot_line.png')
plt.show()
```
